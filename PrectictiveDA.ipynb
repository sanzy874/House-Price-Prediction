```python
acre=df["lot_size_units"]=="acre"
df.loc[acre,"lot_size"]*=43560
df.loc[acre,"lot_size_units"]="sqft"
```

```python
lmean=df["lot_size"].mean()
df["lot_size"].fillna(lmean,inplace=True)
df["lot_size_units"].fillna("sqft",inplace=True)
```
